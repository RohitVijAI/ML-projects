```python
data = dataa
from sklearn.preprocessing import StandardScaler

# Split the data into features (X) and target (y)
X = data.drop('Final', axis=1)
y = data['Final']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Display the scaled features
pd.DataFrame(X_scaled, columns=X.columns).head()
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Initialize and train the Linear Regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

# Predict on the test set
y_pred_linear = linear_reg.predict(X_test)

# Calculate the Mean Squared Error (MSE) for Linear Regression
mse_linear = mean_squared_error(y_test, y_pred_linear)
mse_linear
```

```python
from sklearn.linear_model import Ridge

# Initialize and train the Ridge Regression model
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X_train, y_train)

# Predict on the test set
y_pred_ridge = ridge_reg.predict(X_test)

# Calculate the Mean Squared Error (MSE) for Ridge Regression
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
mse_ridge
```

```python
from sklearn.linear_model import Lasso

# Initialize and train the Lasso Regression model
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X_train, y_train)

# Predict on the test set
y_pred_lasso = lasso_reg.predict(X_test)

# Calculate the Mean Squared Error (MSE) for Lasso Regression
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
mse_lasso
```

```python
from sklearn.linear_model import ElasticNet
    # Initialize the Elastic Net model
elastic_net = ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42)

# Fit the model to the training data
elastic_net.fit(X_train, y_train)

# Make predictions on the test data
y_pred = elastic_net.predict(X_test)

# Evaluate the model
mse_elastic = mean_squared_error(y_test, y_pred)
mse_elastic
```

```python
import numpy as np
import seaborn as sns

# Collect MSE values
mse_values = {
    'Linear Regression': mse_linear,
    'Ridge Regression': mse_ridge,
    'Lasso Regression': mse_lasso,
    'elastic_net Regression' : mse_elastic
}

# Convert to DataFrame for easy plotting
mse_df = pd.DataFrame(list(mse_values.items()), columns=['Model', 'MSE'])

# Plot MSE values for each model
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='MSE', data=mse_df)
plt.title('Mean Squared Error (MSE) for Different Models')
plt.xlabel('Model')
plt.ylabel('Mean Squared Error')
plt.show()
```
