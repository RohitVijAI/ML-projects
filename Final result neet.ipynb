```python
# Calculate basic statistics
max_marks = df['marks'].max()
min_marks = df['marks'].min()
avg_marks = df['marks'].mean()
median_marks = df['marks'].median()
std_dev_marks = df['marks'].std()

print(f"Highest Marks: {max_marks}")
print(f"Lowest Marks: {min_marks}")
print(f"Average Marks: {avg_marks}")
print(f"Median Marks: {median_marks}")
print(f"Standard Deviation: {std_dev_marks}")
```

```python
# Create a DataFrame for the statistics
stats_data = {
    'Statistic': ['Highest Marks', 'Lowest Marks', 'Average Marks', 'Median Marks', 'Standard Deviation'],
    'Value': [max_marks, min_marks, avg_marks, median_marks, std_dev_marks]
}
stats_df = pd.DataFrame(stats_data)

# Plot the statistics
fig, ax = plt.subplots()

stats_df.plot(kind='bar', x='Statistic', y='Value', ax=ax, legend=False, color='skyblue')
plt.title('Basic Statistics of Marks')
plt.ylabel('Marks')
plt.xticks(rotation=45)

# Annotate the bars with the values
for i in ax.containers:
    ax.bar_label(i, label_type='edge')

plt.show()
```

# Understanding the Data:
- Highest Marks (720):-Some students scored perfectly, showing high achievement.
- Lowest Marks (-180):-The lowest score being negative suggests there are penalties for wrong answers, which greatly affect lower scores.
- Average Marks (217.16):-Many students scored below the midpoint since the average score is quite low compared to the highest score.
- Median Marks (163.0):-Half of the students scored below 163, showing that most scores are on the lower end.
- Standard Deviation (166.24):-There's a lot of variation in scores, indicating that students' performance levels are very different from each other.


```python
# Plot distribution
plt.figure(figsize=(10, 6))
plt.hist(df['marks'], bins=10, edgecolor='black', alpha=0.7)

# Add lines for mean, median, min, and max
plt.axvline(avg_marks, color='r', linestyle='dashed', linewidth=1, label=f'Mean: {avg_marks:.2f}')
plt.axvline(median_marks, color='g', linestyle='dashed', linewidth=1, label=f'Median: {median_marks:.2f}')
plt.axvline(min_marks, color='b', linestyle='dashed', linewidth=1, label=f'Min: {min_marks}')
plt.axvline(max_marks, color='y', linestyle='dashed', linewidth=1, label=f'Max: {max_marks}')

# Add labels and title
plt.title('Distribution of NEET Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.legend()
plt.grid(True)
plt.show()
```

- Mean (Red Dashed Line): Shows the average score of all students. It helps you understand the central value of the distribution.
- Median (Green Dashed Line): Indicates the middle score when all students’ scores are arranged in order. This helps to see where the middle of the data lies.
- Minimum (Blue Dashed Line): Marks the lowest score obtained. It shows the leftmost boundary of the score distribution.
- Maximum (Yellow Dashed Line): Represents the highest score obtained. It shows the rightmost boundary of the score distribution.
